```python
# Exercise 6

# Set random seed
torch.manual_seed(1234)

#  Check for access to GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Device:", device)

# Create two random tensors on GPU
tensor_1 = torch.rand(2, 3).to(device)
tensor_2 = torch.rand(2, 3).to(device)

print((tensor_1, tensor_2))
```

```python
# Exercise 7

# Perform matrix multiplication on tensor_1 and tensor_2
tensemult = torch.matmul(tensor_1, tensor_2.T)
tensemult, tensemult.shape
```

```python
# Exercise 8

# Find max
print(f"Maximum value: {tensemult.max()}")

# Find min
print(f"Minimum value: {tensemult.min()}")
```

```python
# Exercise 9

# Find arg max
print(f"Index where max value occurs: {tensemult.argmax()}")

# Find arg min
print(f"Index where min value occurs: {tensemult.argmin()}")
```
